```python
import base64
import urllib.request
import ipywidgets as widgets
from IPython.display import display

url = "https://wikimedia.org"
req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})

try:
    # 1. Download image stream
    with urllib.request.urlopen(req) as response:
        image_data = response.read()
    
    # 2. Convert to Base64
    base64_string = base64.b64encode(image_data).decode('utf-8')
    
    # 3. Create the exact Markdown flag you requested
    markdown_flag = f"![MIT Seal](data:image/svg+xml;base64,{base64_string})"
    
    # 4. Generate a safe, non-freezing text area widget inside Jupyter
    text_box = widgets.Textarea(
        value=markdown_flag,
        placeholder='Markdown code will appear here',
        description='Your Code:',
        disabled=False,
        layout=widgets.Layout(width='100%', height='200px')
    )
    
    print("--- SUCCESS! GENERATED ---")
    print("1. Click inside the text box below.")
    print("2. Press Ctrl+A (or Cmd+A) to highlight everything.")
    print("3. Press Ctrl+C (or Cmd+C) to copy it safely.")
    display(text_box)

except Exception as e:
    print(f"An error occurred: {e}")
    
```
